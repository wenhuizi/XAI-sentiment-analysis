```python
import re
import pandas as pd 
import numpy as np  
import seaborn as sns
import string
import nltk
from nltk.stem.porter import *
import transformers
import shap
```

```python
dataset = pd.read_csv('test.csv')
```

```python
#删除@~用户名
def remove_pattern(input_txt, pattern):
    r = re.findall(pattern, input_txt)
    for i in r:
        input_txt = re.sub(i, '', input_txt)
        
    return input_txt
```

```python
print('\n\nRemoving  Twitter Handles \n\n')
dataset['tidy_text'] = np.vectorize(remove_pattern)(dataset['text'], "@[\w]*")
```

```python
print(dataset['tidy_text'].head())
```

```python
#删除标点符号、数字和特殊字符
dataset['tidy_text'] = dataset['tidy_text'].str.replace("[^a-zA-Z#]", " ")
```

```python
#删除短词
dataset['tidy_text'] = dataset['tidy_text'].apply(lambda x: ' '.join([w for w in x.split() if len(w)>3]))
```

```python
print(dataset['tidy_text'].head())
```

```python
# 缩短字符串以适应管道模型
short_data = [v[:500] for v in dataset["tidy_text"][:20]]
```

```python
#使用pipeline API来快速使用一些预训练模型——情感分析预训练模型：分析文本是正面的还是负面的
classifier = transformers.pipeline("sentiment-analysis", return_all_scores=True)
classifier(short_data[:2])
```

```python
# 将模型输入shap解释器中，创建一个explainer对象，利用它计算每个观察对象的SHAP值，每个特征将对应一个SHAP值
explainer = shap.Explainer(classifier)
```

```python
# explain the predictions of the pipeline on the first two samples
shap_values = explainer(short_data[9:11])
```

```python
shap.plots.text(shap_values[:, :, "POSITIVE"])
```

```python
shap.plots.bar(shap_values[0, :, "POSITIVE"])
```

```python
shap.plots.bar(shap_values[1, :, "NEGATIVE"])
```

```python
shap.plots.bar(shap_values[:, :, "POSITIVE"].mean(0))
```
